```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Criando o DataFrame com os dados meteorolgicos
dados = {
    "Data": ["15/01/2025"] * 5,
    "Cidade": ["So Paulo", "Rio de Janeiro", "Curitiba", "Porto Alegre", "Salvador"],
    "Temperatura Mxima (C)": [30.5, 35.0, 24.0, 28.0, 31.0],
    "Temperatura Mnima (C)": [22.0, 25.0, 18.0, 20.0, 24.5],
    "Precipitao (mm)": [12.0, np.nan, 8.0, 15.0, np.nan],
    "Umidade Relativa (%)": [78, 70, np.nan, 82, 80]
}

df = pd.DataFrame(dados)

# Substituindo valores ausentes
df["Precipitao (mm)"] = df["Precipitao (mm)"].fillna(df["Precipitao (mm)"].mean())
df["Umidade Relativa (%)"] = df["Umidade Relativa (%)"].fillna(df["Umidade Relativa (%)"].median())

# Adicionando a coluna Amplitude Trmica
df["Amplitude Trmica"] = df["Temperatura Mxima (C)"] - df["Temperatura Mnima (C)"]

# Criando um novo DataFrame com cidades cuja Temperatura Mxima  superior a 30C
df_temp_alta = df[df["Temperatura Mxima (C)"] > 30]

# Reordenando as colunas do DataFrame
colunas_ordenadas = ["Data", "Cidade", "Temperatura Mxima (C)", "Temperatura Mnima (C)",
                      "Amplitude Trmica", "Precipitao (mm)", "Umidade Relativa (%)"]
df = df[colunas_ordenadas]

# Criando os dados do grfico de temperatura ao longo do dia
horas = list(range(25))
temperaturas = [15 + (15/12)*h if h <= 12 else 30 - (12/12)*(h-12) for h in horas]

# Gerando o grfico
plt.figure(figsize=(10, 5))
plt.plot(horas, temperaturas, marker='o', linestyle='-', color='b', label="Temperatura (C)")

# Personalizando o grfico
plt.title("Evoluo da Temperatura Durante o Dia")
plt.xlabel("Hora do Dia")
plt.ylabel("Temperatura (C)")
plt.xticks(range(0, 25, 2))  # Marcao de 2 em 2 horas
plt.yticks(range(10, 35, 5))
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Criando o DataFrame com os dados de vendas
dados_vendas = {
    "Dia": ["Segunda", "Tera", "Quarta", "Quinta", "Sexta", "Sbado", "Domingo"],
    "Total de Vendas": [1500, 1800, 2200, 2100, 3000, 4000, 3500],
    "Nmero de Clientes": [120, 140, 160, 150, 200, 250, 230],
    "Lucro": [300, 400, 500, 480, 700, 1000, 900]
}

df = pd.DataFrame(dados_vendas)

# Grfico de Barras - Total de Vendas por Dia
plt.figure(figsize=(10, 5))
sns.barplot(x="Dia", y="Total de Vendas", hue="Dia", data=df, palette="Blues", legend=False)
plt.title("Total de Vendas por Dia")
plt.xlabel("Dia da Semana")
plt.ylabel("Total de Vendas (R$)")
plt.xticks(rotation=45)
plt.show()

# Grfico de Disperso - Nmero de Clientes vs Total de Vendas
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Nmero de Clientes", y="Total de Vendas", data=df, color="b")
plt.title("Relao entre Nmero de Clientes e Total de Vendas")
plt.xlabel("Nmero de Clientes")
plt.ylabel("Total de Vendas (R$)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

# Heatmap - Correlao entre Total de Vendas, Nmero de Clientes e Lucro
plt.figure(figsize=(6, 4))
sns.heatmap(df[["Total de Vendas", "Nmero de Clientes", "Lucro"]].corr(), annot=True, cmap="Blues", fmt=".2f")
plt.title("Matriz de Correlao entre Vendas, Clientes e Lucro")
plt.show()
```
