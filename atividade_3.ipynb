```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Criando o DataFrame com os dados de vendas
dados_vendas = {
    "Dia": ["Segunda", "Terça", "Quarta", "Quinta", "Sexta", "Sábado", "Domingo"],
    "Total de Vendas": [1500, 1800, 2200, 2100, 3000, 4000, 3500],
    "Número de Clientes": [120, 140, 160, 150, 200, 250, 230],
    "Lucro": [300, 400, 500, 480, 700, 1000, 900]
}

df = pd.DataFrame(dados_vendas)

# Gráfico de Barras - Total de Vendas por Dia
plt.figure(figsize=(10, 5))
sns.barplot(x="Dia", y="Total de Vendas", hue="Dia", data=df, palette="Blues", legend=False)
plt.title("Total de Vendas por Dia")
plt.xlabel("Dia da Semana")
plt.ylabel("Total de Vendas (R$)")
plt.xticks(rotation=45)
plt.show()

# Gráfico de Dispersão - Número de Clientes vs Total de Vendas
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Número de Clientes", y="Total de Vendas", data=df, color="b")
plt.title("Relação entre Número de Clientes e Total de Vendas")
plt.xlabel("Número de Clientes")
plt.ylabel("Total de Vendas (R$)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

# Heatmap - Correlação entre Total de Vendas, Número de Clientes e Lucro
plt.figure(figsize=(6, 4))
sns.heatmap(df[["Total de Vendas", "Número de Clientes", "Lucro"]].corr(), annot=True, cmap="Blues", fmt=".2f")
plt.title("Matriz de Correlação entre Vendas, Clientes e Lucro")
plt.show()
```
